```python
# Data Visualization
# Let's create a bar plot to visualize the number of matches per season
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.countplot(data=odi_match_data, x="season")
plt.xticks(rotation=90)
plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
```

```python
# Create a bar plot to visualize the number of matches won by each team
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.countplot(data=odi_match_info, x="winner")
plt.xticks(rotation=90)
plt.title("Number of Matches Won by Each Team")
plt.xlabel("Team")
plt.ylabel("Number of Matches Won")
plt.show()
```

```python
# Set the number of top cities to display
top_n_cities = 10  # You can adjust this number as needed

# Get the top N cities with the highest number of matches
top_cities = odi_match_info['city'].value_counts().nlargest(top_n_cities).index

# Filter the data for the top cities
filtered_odi_match_info = odi_match_info[odi_match_info['city'].isin(top_cities)]

# Create a bar plot to visualize the number of matches in the top cities
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.countplot(data=filtered_odi_match_info, x="city", order=filtered_odi_match_info["city"].value_counts().index)
plt.xticks(rotation=90)
plt.title(f"Top {top_n_cities} Cities with the Most Matches (Match Info Data)")
plt.xlabel("City")
plt.ylabel("Number of Matches")
plt.show()
```

# Filter match info data for Bangladesh matches

```python
# Filter match info data for Bangladesh matches
bangladesh_matches_info = odi_match_info[(odi_match_info['team1'] == 'Bangladesh') | (odi_match_info['team2'] == 'Bangladesh')]

# Filter match data for Bangladesh matches
bangladesh_matches_data = odi_match_data[odi_match_data['match_id'].isin(bangladesh_matches_info['id'])]
```

# Bangladesh's Performance Over Seasons

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.lineplot(data=bangladesh_matches_info, x="season", y="win_by_runs", label="Win by Runs")
sns.lineplot(data=bangladesh_matches_info, x="season", y="win_by_wickets", label="Win by Wickets")
plt.xticks(rotation=90)
plt.title("Bangladesh's Performance Over Seasons (Win by Runs/Wickets)")
plt.xlabel("Season")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Calculate win counts for each team
win_counts = odi_match_info['winner'].value_counts()
```

```python
# Create a bar plot for win-loss analysis
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x=win_counts.index, y=win_counts.values)
plt.xticks(rotation=90)
plt.title("Win-Loss Analysis of ODI Cricket Matches (2002-2023)")
plt.xlabel("Team")
plt.ylabel("Number of Matches Won")
plt.show()
```

```python
# Calculate win percentage for each team
total_matches_played = pd.concat([odi_match_info['team1'], odi_match_info['team2']]).value_counts()
win_percentage = (win_counts / total_matches_played) * 100
win_percentage = win_percentage.round(2)
```

```python
# Create a bar plot for win percentage analysis
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x=win_percentage.index, y=win_percentage.values)
plt.xticks(rotation=90)
plt.title("Win Percentage Analysis of ODI Cricket Matches (2002-2023)")
plt.xlabel("Team")
plt.ylabel("Win Percentage")
plt.ylim(0, 100)  # Set the y-axis limit to 0-100%
plt.show()
```

```python
odi_match_info['date'] = pd.to_datetime(odi_match_info['date'])
```

```python
# Group the data by date
date_grouped = odi_match_info.groupby('date')

# Calculate the number of matches played on each date
matches_played_per_date = date_grouped.size()

# Calculate the number of matches won by each team on each date
matches_won_per_date = date_grouped['winner'].value_counts().unstack(fill_value=0)
```

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
matches_played_per_date.plot()
plt.title("Matches Played Over Time (2002-2023)")
plt.xlabel("Date")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
```

# Batting and bowling averages for Bangladesh

```python
# Calculate batting and bowling averages for Bangladesh
batting_average = bangladesh_matches_info.groupby('season')['win_by_runs'].mean()
bowling_average = bangladesh_matches_info.groupby('season')['win_by_wickets'].mean()

# Plot batting and bowling averages over the years
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
plt.plot(batting_average.index, batting_average.values, marker='o', label='Batting Average')
plt.plot(bowling_average.index, bowling_average.values, marker='o', label='Bowling Average')
plt.title("Bangladesh's Batting and Bowling Averages (2002-2023)")
plt.xlabel("Season")
plt.ylabel("Average")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# The top players of the match for Bangladesh


```python
# Find the top players of the match for Bangladesh
top_players =bangladesh_matches_info['player_of_match'].value_counts().nlargest(5)

# Plot the top players of the match for Bangladesh
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x=top_players.index, y=top_players.values)
plt.title("Top Players of the Match for Bangladesh")
plt.xlabel("Player")
plt.ylabel("Number of Times Player of the Match")
plt.xticks(rotation=45)
plt.show()
```

# The percentage of matches won when Bangladesh won the toss and when they lost the toss

```python
# Calculate the percentage of matches won when Bangladesh won the toss and when they lost the toss
matches_won_when_won_toss = bangladesh_matches_info[bangladesh_matches_info['toss_winner'] == 'Bangladesh']
matches_won_when_lost_toss = bangladesh_matches_info[bangladesh_matches_info['toss_winner'] != 'Bangladesh']

win_percentage_when_won_toss = (matches_won_when_won_toss['winner'].value_counts() / len(matches_won_when_won_toss)) * 100
win_percentage_when_lost_toss = (matches_won_when_lost_toss['winner'].value_counts() / len(matches_won_when_lost_toss)) * 100

# Plot the results
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
plt.bar(win_percentage_when_won_toss.index, win_percentage_when_won_toss.values, label='Winning When Won Toss')
plt.bar(win_percentage_when_lost_toss.index, win_percentage_when_lost_toss.values, label='Winning When Lost Toss')
plt.title("Impact of Toss on Bangladesh's Match Results")
plt.xlabel("Opponent Team")
plt.ylabel("Winning Percentage")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# Winning percentage at each venue for Bangladesh


```python
# Calculate winning percentage at each venue
venue_winning_percentage = (bangladesh_matches_info['venue'].value_counts() / len(bangladesh_matches_info)) * 100

# Plot the top venues for Bangladesh
top_venues = venue_winning_percentage.nlargest(10)
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x=top_venues.values, y=top_venues.index, orient="h")
plt.title("Top Venues for Bangladesh (by Winning Percentage)")
plt.xlabel("Winning Percentage")
plt.ylabel("Venue")
plt.show()
```
